````python
msigdb = dc.get_resource('MSigDB')
msigdb
````

````python
msigdb['collection'].unique()
````

````python
# Filter by hallmark
msigdb = msigdb[msigdb['collection']=='go_biological_process']

# Remove duplicated entries
msigdb = msigdb[~msigdb.duplicated(['geneset', 'genesymbol'])]

msigdb
````

````python
# Rename
msigdb.loc[:, 'geneset'] = [name.split('GOBP_')[1] for name in msigdb['geneset']]
msigdb
````

````python
gene_tables = {}
for i in range(0,4):
    gene_tables[f'Q{i+1}'] = pd.read_csv(f'Q{i+1}_table_lm_sig.txt', sep='\t', index_col=0)
````

````python
# Infer enrichment with ora using significant deg
go_pvals = {}
for i in range(0,4):
    top_genes = gene_tables[f'Q{i+1}'].copy()
    top_genes = top_genes.set_index('gene_name')
    top_genes['group'] = 'RPS2.vs.STAMP'
    
    # Run ora
    enr_pvals = dc.get_ora_df(
        df=top_genes,
        net=msigdb,
        source='geneset',
        target='genesymbol'
    )
    
    enr_pvals = enr_pvals[enr_pvals['FDR p-value']<=0.05]
    enr_pvals['-log10 FDR p-value'] = -np.log10(enr_pvals['FDR p-value'])
    go_pvals[f'Q{i+1}'] = enr_pvals
````

````python
# Set 0s to min p-value
sns.set(rc={'figure.figsize':(3,5)})
sns.set_style('whitegrid')
for i in range(0,4):
    go_pvals[f'Q{i+1}'][['Term', '-log10 FDR p-value', 'Odds ratio']].sort_values(by='-log10 FDR p-value', ascending=False).head(10).plot.scatter(x='-log10 FDR p-value',
                      y='Term',
                      s='Odds ratio',
                      color='black')
    plt.xticks(rotation=90)
    plt.title(f'Q{i+1}')
    plt.show()
````
